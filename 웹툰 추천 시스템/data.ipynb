```python
# 저자 정보를 분리하여 새로운 컬럼 생성
def split_authors(row):
    authors = row['author'].split('/')
    num_authors = len(authors)

    # "/"로 구분되지 않은 경우 (글작가, 그림작가, 원작자 동일 인물)
    if num_authors == 1:
        return pd.Series([authors[0], authors[0], authors[0]], index=['writer', 'illustrator', 'original_creator'])
    # "/"로 구분된 경우가 2개인 경우 (글작가와 원작자는 첫번째 사람, 그림작가는 두번째 사람)
    elif num_authors == 2:
        return pd.Series([authors[0], authors[1], authors[0]], index=['writer', 'illustrator', 'original_creator'])
    # "/"로 구분된 경우가 3개인 경우 (순서대로 글작가, 그림작가, 원작자)
    elif num_authors == 3:
        return pd.Series(authors, index=['writer', 'illustrator', 'original_creator'])

# data[['writer', 'illustrator', 'original_creator']] = data.apply(split_authors, axis=1)
# data
```
